Lab 3: Chi-Square for independence test

```python
# Step1: Import required libraries
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# Step2: Create the observed frequency (cintigency) table
# Example: gender vs Product Preference
data = {
    "Like": [30,20],
    "Dislike": [10,40]
}
```

```python
observed = pd.DataFrame(data, index=["Male", "Female"])
print("Observed Frequency Table:")
print(observed)
```

```python
chi2_stat, p_value, dof, expected = chi2_contingency(observed)
```

```python
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("P-value:", p_value)

print("\nExpected Frequency Table:")
print(pd.DataFrame(expected, index=observed.index, columns=observed.columns))
```

```python
# Step5: Hypothesis Decision
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis. There is a significant association between gender and product preference.")
else:
  print("\nFail to reject the null hypothesis. There is no significant association between gender and product preference.")
```
